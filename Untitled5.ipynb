```python
high_delay_regions <- sqldf("
SELECT home_zone,
       COUNT(*) AS total_delays
FROM combined_data
WHERE complaint_type = 'Delay'
GROUP BY home_zone
ORDER BY total_delays DESC
")

print(high_delay_regions)
```

```python
customer_complaints <- sqldf("
SELECT customer_type,
       complaint_type,
       COUNT(*) AS total_complaints
FROM combined_data
GROUP BY customer_type, complaint_type
ORDER BY total_complaints DESC
")

print(customer_complaints)
```

```python
import pandas as pd
```

```python
customers = pd.read_csv("customers.csv")
complaints = pd.read_csv("complaints.csv")
```

```python
combined_data = pd.merge(customers,
                         complaints,
                         on="customer_id")

print(combined_data.head())
```

```python
combined_data.isnull().sum()
```

```python
combined_data["loyalty_score"] = combined_data["loyalty_score"].fillna(
    combined_data["loyalty_score"].mean()
)

combined_data["compensation_amount"] = combined_data["compensation_amount"].fillna(0)

combined_data["preferred_channel"] = combined_data["preferred_channel"].fillna("Unknown")

combined_data.isnull().sum()
```

```python
combined_data["home_zone"] = combined_data["home_zone"].str.upper()

combined_data["home_zone"].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))

sns.countplot(
    data=combined_data,
    x="home_zone"
)

plt.title("Complaint Distribution by Operational Region")
plt.xlabel("Operational Region")
plt.ylabel("Number of Complaints")

plt.show()
```

```python
combined_data[[
    "loyalty_score",
    "resolution_days",
    "compensation_amount"
]].corr()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=combined_data,
    x="severity",
    order=combined_data["severity"].value_counts().index
)

plt.title("Distribution of Complaint Severity Levels")
plt.xlabel("Severity Level")
plt.ylabel("Number of Complaints")

plt.show()
```
